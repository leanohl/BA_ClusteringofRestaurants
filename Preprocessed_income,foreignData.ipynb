```python
import pandas as pd
import numpy as np
```

```python
#import data
df= pd.read_csv('data/KoreaIncomeData.csv') 
df.head()  # 40441 rows and 199 columns
```

```python
#We only use 'Gu' and 'Income' data.
Korea_income=df.copy()[['GU','ADQ3E']]

#Change the column names
Korea_income.columns = ['Gu','Income']

#Drop null data and we have 27443 rows and 2 columns
nullIndex=Korea_income[(Korea_income['Income']=='#NULL!')].index
Korea_income.drop(nullIndex,inplace=True)
```

```python
#About Gu information
Korea_income['Gu']=Korea_income.Gu.replace({110:"Jongno-gu",140:"Jung-gu",170:"Yongsan-gu",200:"Seongdong-gu",215:"Gwangjin-gu",230:"Dongdaemun-gu",260:"Jungnang-gu",290:"Seongbuk-gu",305:"Gangbuk-gu",320:"Dobong-gu",350:"Nowon-gu",380:"Eunpyeong-gu",410:"Seodaemun-gu",440:"Mapo-gu",470:"Yangcheon-gu",500:"Gangseo-gu",530:"Guro-gu",545:"Geumcheon-gu",560:"Yeongdeungpo-gu",590:"Dongjak-gu",620:"Gwanak-gu",650:"Seocho-gu",680:"Gangnam-gu",710:"Songpa-gu",740:"Gangdong-gu"})
```

```python
#About Income information
Korea_income['Income']=Korea_income.Income.replace({'1':"Under 100",'2':"Under 100",'3':"100~200",'4':"100~200",'5':"200~300",'6':"200~300",'7':"300~400",'8':"300~400",'9':"400~500",'10':"400~500",'11':"500~600",'12':"500~600",'13':"600~700",'14':"600~700",'15':"700~800",'16':"700~800",'17':"800~900",'18':"800~900",'19':"900~1000",'20':'900~1000','21':"Over 1000"})
```

```python
Korea_income.reset_index(drop=True,inplace=True)
Korea_income.head()
```
